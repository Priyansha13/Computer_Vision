```python
import cv2

img = cv2.imread("img1.jpg")  # Check image path

# Split and display individual channels
b, g, r = cv2.split(img) 
cv2.imshow("Blue Channel", b)
cv2.imshow("Green Channel", g)
cv2.imshow("Red Channel", r)

cv2.waitKey(0)  # Wait for a key press
```

```python
# Use original image directly
cv2.imshow("Original Image", img)

cv2.waitKey(0)  # Wait for a key press
```

```python
# Merge split channels back together
img1 = cv2.merge((b, g, r))  
cv2.imshow("Merged Image", img1)

cv2.waitKey(0)  # Wait for a key press
```

- ii)	Convert the given images in to different formats (For ex: *.jpeg, *.jpg, *.jpe, , *.png, tiff, *.tif, e.t.c)

```python
# To .tiff format
input_file = "img1.jpg"
output_file = "image.tiff"
img = Image.open('img1.jpg')
img.save('image.tiff', "TIFF")
print(f"Image converted from {input_file} to {output_file} (TIFF)")
```

```python
# To .jpe format
import os

input_file = "img1.jpg"
output_file = os.path.splitext(input_file)[0] + ".jpe"  # Change extension to .jpe

os.rename(input_file, output_file)
print(f"Successfully renamed '{input_file}' to '{output_file}'.")
```

```python
img = Image.open(output_file)
plt.imshow(img)
plt.title("Renamed Image (.jpe)")
plt.show()
```

```python
input_file = "img1.jpe"
output_file_png = os.path.splitext('img1.jpg')[0] + ".png"  # Change extension to .png

img = Image.open(input_file)
img.save(output_file_png, "PNG")  
print(f"Image converted from '{input_file}' to '{output_file_png}' (PNG)")
```

```python
img_png = Image.open(output_file_png)
plt.imshow(img_png)
plt.title("Converted Image (.png)")
plt.show()
```
