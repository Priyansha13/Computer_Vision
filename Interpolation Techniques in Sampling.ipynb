```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
```

```python
# Read the image in grayscale
image = plt.imread('img4.jpg')
gray_image = np.dot(image[..., :3], [0.2989, 0.5870, 0.1140])  # Convert to grayscale
```

```python
# Define interpolation functions

def bilinear_interpolation(image, new_shape):
    return ndimage.zoom(image, (new_shape[0] / image.shape[0], new_shape[1] / image.shape[1]), order=1)

def bicubic_interpolation(image, new_shape):
    return ndimage.zoom(image, (new_shape[0] / image.shape[0], new_shape[1] / image.shape[1]), order=3)

def nearest_neighbor_interpolation(image, new_shape):
    return ndimage.zoom(image, (new_shape[0] / image.shape[0], new_shape[1] / image.shape[1]), order=0)

def lanczos4_interpolation(image, new_shape):
    return ndimage.zoom(image, (new_shape[0] / image.shape[0], new_shape[1] / image.shape[1]), order=4)
```

```python
# Define the desired new shape with a larger zoom factor
zoom_factor = 10
new_shape = (image.shape[0] * zoom_factor, image.shape[1] * zoom_factor)
```

```python
# Perform interpolation using different methods
bilinear_result = bilinear_interpolation(gray_image, new_shape)
bicubic_result = bicubic_interpolation(gray_image, new_shape)
nearest_neighbor_result = nearest_neighbor_interpolation(gray_image, new_shape)
lanczos4_result = lanczos4_interpolation(gray_image, new_shape)
```

```python
# Display the results
plt.figure(figsize=(15, 15))

plt.subplot(231)
plt.title('Original Grayscale')
plt.imshow(gray_image, cmap='gray')
plt.axis('off')

plt.subplot(232)
plt.title('Bilinear Interpolation')
plt.imshow(bilinear_result, cmap='gray')
plt.axis('off')

plt.subplot(233)
plt.title('Bicubic Interpolation')
plt.imshow(bicubic_result, cmap='gray')
plt.axis('off')

plt.subplot(234)
plt.title('Nearest Neighbor Interpolation')
plt.imshow(nearest_neighbor_result, cmap='gray')
plt.axis('off')

plt.subplot(235)
plt.title('Lanczos4 Interpolation')
plt.imshow(lanczos4_result, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
